```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#reading the dataset using pandas lib
#selecting only col1, and item

dataset = pd.read_csv('cacSurveys3May2016-withColumnNames.csv',usecols=["Column 1","Item"])
dataset.head()
```

```python
#only selected rows are stored in the .csv file to get the file
dataset.to_csv('generalshop.csv')



##reading the file from actual dataset 
actual_dataset=pd.read_csv('generalshop.csv',usecols=["Column 1","Item"])
actual_dataset.head()
```

```python
#calulaiton of the frequency of each item ...
#item{key:value}

Items={}
cnt=0
for item in actual_dataset['Item']:
    if item in Items:
        Items[item]=Items[item]+1
    else:
        Items[item]=1
        cnt+=1
```

```python
Items
```

```python
keys=[]
vals=[]
for i,k in Items.items():
        keys.append(i)
        vals.append(k)
```

```python
plt.bar(keys, vals, label="Items sold")
plt.rcParams["figure.figsize"] = [20,20]
plt.ylabel ('Number of Transactions')
plt.xlabel ('Items Sold')
plt.xticks(list(keys), rotation=90)
plt.legend (bbox_to_anchor=(1, 1), loc="best", borderaxespad=0.)
plt.show()
```
